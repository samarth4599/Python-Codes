```python
def rot(a):
    t=a[0]
    for i in range(len(a)-1):
        a[i]=a[i+1]
    a[len(a)-1]=t
def times(a,k):
    for i in range(k):
        rot(a)
a=[1,2,3,4]
times(a,2)
print(a)
```

```python
#sum of digits
def digital_root(n):
    return n if n < 10 else digital_root(sum(map(int,str(n))))
```

```python
#convert string to camel case
def to_camel_case(text):
    removed = text.replace('-', ' ').replace('_', ' ').split()
    if len(removed) == 0:
        return ''
    return removed[0]+ ''.join([x.capitalize() for x in removed[1:]])
```

```python
#Delete occurrences of an element if it occurs more than n times
def delete_nth(order,max_e):
    ans = []
    for o in order:
        if ans.count(o) < max_e: ans.append(o)
    return ans
```

### Merge Sort

```python
a=[int(input()) for i in range(int(input("Enter no of elements")))]
l=len(a)
def merge(a,start,end):
    if start<end:
        mid=(start+end)//2
        merge(a,start,mid)
        merge(a,mid+1,end)
        merging(a,start,end)
def merging(a,start,end):
    p=start
    mid=(start+end)//2
    q=mid
    f=[]
    j=0
    while(p<mid and q<end):
        if a[p]<a[q]:
            f.append(a[p])
            p+=1
        else:
            f.append(a[q])
            q+=1
    while(p<mid):
        f.append(a[p])
        p+=1
    while(q<end):
        f.append(a[q])
        q+=1
    for i in range(len(f)):
        a[start+i]=f[i]
merge(a,0,l)
print("The sorted list is \n",a)
```

### Quick Sort

```python
a=[int(input()) for i in range(int(input("Enter no of elements")))]
l=len(a)
def quickSort(a,start,end):
    if start<end:
        pp=partition(a,start,end)
        quickSort(a,start,pp-1)
        quickSort(a,pp+1,end)
def partition(a,start,end):
    pivot=a[start]
    count,i = 0,start+1
    while(i!=end+1):
        if a[i]<pivot:
            count+=1
        i+=1
    pi=start+count
    a[pi],a[start]=a[start],a[pi]
    i,j = start,end
    while i<j:
        while i<=end and a[i]<=pivot:
            i+=1
        while a[j]>pivot:
            j-=1
        if i<=j:
            a[i],a[j]=a[j],a[i]
            i+=1
            j-=1
    return pi
quickSort(a,0,l-1)
print("The sorted list is \n",a)
```

### Binary Search

```python
a=[int(input("Enter element")) for i in range(int(input("Enter no of elements--in sorted order")))]
val=int(input("Enter value to search"))
def binsort(a,val):
    n=len(a)
    front=0
    back=n-1
    if val<a[front] or val>a[back]:
        return -1
    while front<=back:
        mid=(front+back)//2
        if val==a[mid]:
            return mid
        elif val>a[mid]:
            front=mid+1
        else:
            back=mid-1
    return -1
index=binsort(a,val)
print("The index is    ",index)
```

### Selection Sort

```python
a=[int(input("Enter element")) for i in range(int(input("Enter no of elements")))]
def selsort(a):
    for i in range(len(a)-1):
        minn=a[i]
        minix=i
        for j in range(i+1,len(a)):
            if minn>a[j]:
                minn=a[j]
                minix=j
        if minix!=i:
            a[i],a[minix]=a[minix],a[i]
    print(a)
selsort(a)
```
